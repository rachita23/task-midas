Parse the JSON file to display the following content for each tweet:
1. FULL TEXT
2. DATE & TIME
3. NUMBER OF RETWEETS
4. NUMBER OF FAVOURITES/LIKES
5. NUMBER OF IMAGES

```python
#To show the ouput in tabular form I chose dataframe
df = pd.DataFrame({'FULL TEXT': [], 'DATE & TIME': [], '#RETWEETS': [], '#LIKES': [], '#IMAGES': []})
for i in range(len(tweets_json)):
    #Check if the given tweet contains images or not
    if 'media' in tweets_json[i]['entities']:
        img = len(tweets_json[i]['entities']['media'])
    else:
        img = "None"
    df = df.append({'FULL TEXT': tweets_json[i]['full_text'], 'DATE & TIME': tweets_json[i]['created_at'].split("+")[0], '#RETWEETS': int(tweets_json[i]['retweet_count']), '#LIKES': int(tweets_json[i]['favorite_count']), '#IMAGES': img}, ignore_index=True)
```
